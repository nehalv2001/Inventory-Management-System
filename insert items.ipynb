```python
import json

#read json file
file=open('record.json','r')
r=file.read()
record=json.loads(r)

#take input for new item
product_name=input('enter product name : ')
price=int(input('enter price : '))
quantity=int(input('enter quantity : '))
company=input('enter company name  : ')
expire=input('enter expiry date : ')

#update the record
record[len(record)+1]={'name':product_name,'price':price,'quantity':quantity,'company':company,'expire':expire}

#write updated record in json file
file=open('record.json','w')
js=json.dumps(record)
file.write(js)
file.close()

#view updated record
file=open('record.json','r')
r=file.read()
record=json.loads(r)
record
```
