```python
import json

file=open('record.json','r')
js=file.read()
record=json.loads(js)
file.close()

file=open('sales.json','r')
js=file.read()
sales=json.loads(js)
file.close()


for rec in record:
    print('id:',rec,'|| name:',record[rec]['name'],'|| price:',record[rec]['price'],'|| quantity:',record[rec]['quantity'])
print('============================================================')   
print('select item from above list')
id=int(input('enter id of item : '))
quantity=int(input('enter quantity of item : '))

try:
    if quantity>record[str(id)]['quantity']:
        print('!!!! enter valid quantity !!!!')
    else:
        print('=======================================')
        print('-----------------BILL------------------')
        print('product id :',id)
        print('product name :',record[str(id)]['name'])
        print('product price :',record[str(id)]['price'])
        print('product quantity :',quantity)
        print('---------------------------------------')
        print('total amount :',record[str(id)]['price']*quantity)
        print('=======================================')
        
        file=open('record.json','w')
        record[str(id)]['quantity']=record[str(id)]['quantity']-quantity
        js=json.dumps(record)
        file.write(js)
        file.close()
        
        file=open('sales.json','w')
        sales[str(len(sales)+1)]={'name':record[str(id)]['name'],'price':record[str(id)]['price'],'quantity':quantity,'total':record[str(id)]['price']*quantity}
        js=json.dumps(sales)
        file.write(js)
        file.close()
        
except:
    print('!!!! your item not found !!!!')


```
